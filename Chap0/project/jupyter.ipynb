## 理解编程
编程,就是用指令安排电脑做具体的事。
程序就是指令。由一条条指令组成。

为了更省力,所以设计了各种编程言语。 比如:Python
一系列指令构成一个功能。(比如:读取一个文档)
这样就可以总减少重复劳动,功能可以重复调用。

就像是将军发号司令,下级分解任务并执行。

## 程序
*任何复杂的东西都是由基本的,简单的东西构成。*
语言是由句子构成,句子是由词组构成,词组由单字构成。单字由笔划(汉字)或字母(英文)构成。

复杂的程序也是由简单的,基本的程序语句构成。

学习编程,就是从最基本的的语句开始。
与语言相比较,编程基本语句少得可怜。所需记忆量并不大。

## 如何学会
编程的难点在于编程是符号化的代码。比较抽象。
需要熟练操作才能运用自如。也就是说要足够多的练习。
"熟能生巧"
